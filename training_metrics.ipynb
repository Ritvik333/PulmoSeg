```python
import pandas as pd
import matplotlib.pyplot as plt

# List your CSV file paths with full /content/ prefix
csv_files = [
    '/content/AAGS_training_log.csv',    # Adjust file names exactly as in your /content directory
    '/content/UNetCBAM_training_log.csv',
    '/content/UNet_training_log.csv'
]

plt.figure(figsize=(10, 6))

for file in csv_files:
    # Load CSV from file
    df = pd.read_csv(file)

    # Extract model name from filename for legend label (remove path and extension)
    model_name = file.split('/')[-1].replace('_training_log.csv', '')

    # Plot Average Accuracy over epochs
    plt.plot(df['Epoch'], df['Average_Accuracy'], marker='o', label=model_name)

plt.title('Training Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Average Accuracy (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List your CSV file paths with full /content/ prefix
csv_files = [
    '/content/AAGS_training_log.csv',
    '/content/UNetCBAM_training_log.csv',
    '/content/UNet_training_log.csv'
]

plt.figure(figsize=(10, 6))

for file in csv_files:
    # Load CSV from file
    df = pd.read_csv(file)

    # Extract model name from filename for legend label
    model_name = file.split('/')[-1].replace('_training_log.csv', '')

    # Plot Average Loss over epochs
    plt.plot(df['Epoch'], df['Average_Loss'], marker='o', label=model_name)

plt.title('Training Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Average Loss')
plt.legend()
plt.grid(True)
plt.show()
```
